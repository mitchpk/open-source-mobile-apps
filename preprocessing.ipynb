```python
import re
import nltk
import pycld2 as cld2
nltk.download('stopwords')
nltk.download('wordnet')

lst_stopwords = nltk.corpus.stopwords.words("english")
lem = nltk.stem.wordnet.WordNetLemmatizer()

def preprocess(text: str):
    text = re.sub(r'[^\w\s]', '', text.lower().strip())
    lst_text = text.split()
    lst_text = [word for word in lst_text if word not in lst_stopwords]
    lst_text = [lem.lemmatize(word) for word in lst_text]
    text = " ".join(lst_text)
    return text

def is_english(data: pd.Series):
    try:
        isReliable, _, details = cld2.detect(data["content"])
        if isReliable and details[0][0] == "ENGLISH":
            return True
    except:
        pass
    
    return False

def process_reviews(reviews: list):
    result = []
    for review in reviews:
        if review["content"] and len(review["content"]) > 100:
            if is_english(review):
                result.append(preprocess(review["content"]))
            
    return result

#reviews_processed = {}
#for k in reviews:
#    print(k)
#    reviews_processed[k] = process_reviews(reviews[k])
```
