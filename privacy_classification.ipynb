```python
all_privacy = reviews_df[(reviews_df['label'] == True) & (reviews_df['content'].str.len() > 400)].copy()
```

```python
import spacy
nlp = spacy.load('en_core_web_md')

i = 0
print(len(all_privacy))

def remove_orgs(text: str):
    global i
    i += 1
    print(i, end="\r")
    doc = nlp(text)
    newString = text
    for e in reversed(doc.ents):
        if e.label_ == "ORG": # Only redact PERSON entities
            start = e.start_char
            end = start + len(e.text)
            newString = newString[:start] + "xxxx" + newString[end:]
    return newString

all_privacy['content'] = all_privacy['content'].apply(remove_orgs)
```

```python
genres = ["Communication", "Productivity", "Tools"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import Normalizer
from sklearn.cluster import KMeans
import scipy.spatial.distance as scdist
from sklearn.metrics import silhouette_score, silhouette_samples

for i in genres:
    apps = pd.concat([os_apps[os_apps['genre'] == i], cs_apps[cs_apps['genre'] == i]])
    reviews = all_privacy[all_privacy['appId'].isin(apps['appId'])]
    
    vectorizer = TfidfVectorizer(
        stop_words="english"
    )
    t0 = time()
    X_tfidf = vectorizer.fit_transform(reviews['content'])

    print(f"vectorization done in {time() - t0:.3f} s")
    print(f"n_samples: {X_tfidf.shape[0]}, n_features: {X_tfidf.shape[1]}")
    print(f"{X_tfidf.nnz / np.prod(X_tfidf.shape):.3f}")

    lsa = make_pipeline(TruncatedSVD(n_components=100, random_state=42), Normalizer(copy=False))
    t0 = time()
    X_lsa = lsa.fit_transform(X_tfidf)
    explained_variance = lsa[0].explained_variance_ratio_.sum()

    print(f"LSA done in {time() - t0:.3f} s")
    print(f"Explained variance of the SVD step: {explained_variance * 100:.1f}%")
    
    k_range = range(2,11)
    kms = [KMeans(n_clusters=k, init='k-means++', n_init=1, random_state=42).fit(X_lsa) for k in k_range]
    centroids = [X.cluster_centers_ for X in kms]
    labels = [km.labels_ for km in kms]

    #calculate Euclidean distance from each point to cluster center
    k_euclid = [scdist.cdist(X_lsa, c, 'euclidean') for c in centroids] 
    dist = [np.min(ke, axis=1) for ke in k_euclid]
    #Total within cluster sum of squares
    wcss = [sum(d**2) for d in dist]
    #average wcss
    avwcss = [(sum(d**2))/len(d) for d in dist]
    #total sum of squares
    #tss = sum(scdist.pdist(X_lsa)**2)/X_lsa.shape[0]
    #between cluster sum of squares:
    #bss = tss - wcss
    #plot average wcss vs number of clusters "Elbow plot": look for a point where the rate of decrease in wcss sharply shifts
    #plt.subplots(figsize=(18, 12)) # set size
    plt.plot(k_range, avwcss, '-o')
    plt.title(i)
    plt.ylabel("average wcss")
    plt.xlabel("k")
    plt.show()
    
    silhouette_avg_scores = [silhouette_score(X_lsa, l) for l in labels]
    plt.plot(k_range, silhouette_avg_scores, '-o')
    plt.title(i)
    plt.ylabel("silhouette averages")
    plt.xlabel("k")
    plt.show()
```

```python
from sklearn.manifold import TSNE
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sb
import matplotlib.patches as mpatches

def scatter(x, labels):
    n_labels = len(set(labels))
    palette = np.array(sb.color_palette("hls", n_labels )) # color palette with seaborn.
    f = plt.figure(figsize=(18, 12))
    ax = plt.subplot(aspect='equal')
    sc = ax.scatter(x[:,0], x[:,1], linewidth=0, s=40, color=palette[labels])
    #ax.axis('off')
    ax.axis('tight')
    txts = []
    color_patches = []
    for i in range(n_labels):
        # Position of each label
        patch = mpatches.Patch(color=palette[i], label= "Cluster %i "% i)
        color_patches.append(patch)
        xtext, ytext = np.median(x[labels == i, :], axis=0)
        txt = ax.text(xtext, ytext, str(i), fontsize=18)
        txts.append(txt)
    plt.legend(handles=color_patches, bbox_to_anchor=(1.04,1), loc="upper left", fontsize=14)
    plt.subplots_adjust(right=0.60)
    plt.show()

ks = [9, 8, 6]
clusters = {}
for i in range(len(genres)):
    k = ks[i]
    genre = genres[i]
    print(genre, "with k =", k)
    apps = pd.concat([os_apps[os_apps['genre'] == genre], cs_apps[cs_apps['genre'] == genre]])
    reviews = all_privacy[all_privacy['appId'].isin(apps['appId'])]
    
    vectorizer = TfidfVectorizer(
        stop_words="english"
    )
    t0 = time()
    X_tfidf = vectorizer.fit_transform(reviews['content'])

    print(f"vectorization done in {time() - t0:.3f} s")
    print(f"n_samples: {X_tfidf.shape[0]}, n_features: {X_tfidf.shape[1]}")
    print(f"{X_tfidf.nnz / np.prod(X_tfidf.shape):.3f}")

    lsa = make_pipeline(TruncatedSVD(n_components=100, random_state=42), Normalizer(copy=False))
    t0 = time()
    X_lsa = lsa.fit_transform(X_tfidf)
    explained_variance = lsa[0].explained_variance_ratio_.sum()

    print(f"LSA done in {time() - t0:.3f} s")
    print(f"Explained variance of the SVD step: {explained_variance * 100:.1f}%")
    
    kmeans = KMeans(
        n_clusters=k,
        max_iter=300,
        n_init=1,
        random_state=None,
    ).fit(X_lsa)
    cluster_ids, cluster_sizes = np.unique(kmeans.labels_, return_counts=True)
    print(f"Number of elements asigned to each cluster: {cluster_sizes}")
    
    original_space_centroids = lsa[0].inverse_transform(kmeans.cluster_centers_)
    order_centroids = original_space_centroids.argsort()[:, ::-1]

    tsne_cos = TSNE(n_components=2, 
                    perplexity=30, 
                    learning_rate=1000, 
                    n_iter=1000, 
                    metric='cosine', 
                    verbose=1)
    dist = 1 - cosine_similarity(X_lsa)

    tsne_cos_coords = tsne_cos.fit_transform(dist)

    scatter(tsne_cos_coords, kmeans.labels_)
    clustered = reviews.copy()
    clustered['cluster'] = kmeans.labels_
    clustered['silhouette'] = silhouette_samples(X_lsa, clustered['cluster'])
    clusters[genre] = clustered
    ax = clustered.groupby('cluster')['silhouette'].hist(bins=20, alpha=0.5, legend=True)
    plt.show()
```

```python
def print_influential(data: pd.DataFrame, filter_clusters):
    sorted_ = data.groupby('cluster').apply(lambda x: x.sort_values('silhouette', ascending=False))
    for i in filter_clusters:
        print("Cluster", i)
        stuff = sorted_.loc[i]
        print("Proportion of open-source apps in this cluster:", len(stuff[stuff['appId'].isin(os_apps['appId'])])/len(stuff))
        print("---------")
        for j in range(0, 10):
            print(sorted_.loc[i].iloc[j]['content'])
            print()

for k, v in clusters.items():
    print(k)
    filter_clusters = v.groupby('cluster')['silhouette'].mean().nlargest(3).keys().tolist()
    print("Primary clusters:", filter_clusters)
    open_source = v['appId'].isin(os_apps['appId'])
    print("Open source")
    v[open_source].groupby('cluster')['silhouette'].hist(bins=20, alpha=0.5, legend=True)
    plt.show()
    print("Closed source")
    v[~open_source].groupby('cluster')['silhouette'].hist(bins=20, alpha=0.5, legend=True)
    plt.show()
    print_influential(v, filter_clusters)
    num_os = len(v[open_source & v['cluster'].isin(filter_clusters)])
    num_cs = len(v[~open_source & v['cluster'].isin(filter_clusters)])
    print("Total proportion of open-source apps in primary clusters:", num_os/len(v[open_source]))
    print("Total proportion of closed-source apps in primary clusters:", num_cs/len(v[~open_source]))
    print()
```
